```python
""" Given a list of numbers ,i.e., [1,2,4,-5,7,9,3,2], make another list that contains all the items in sorted order from min to max ,i.e., your result will be another list like [-5,1,2,2,3,4,7,9]"""
def findMin(L,startIdx):
    minval = L[startIdx]
    idx = startIdx    
    for i in range(startIdx, len(L)):
        if L[i]<minval:            
            minval = L[i]
            idx = i
        else:
            pass
    return minval,idx

def swapVal(L,i1,i2):
    temp = L[i1]
    L[i1] = L[i2]
    L[i2] = temp
    return L
    
from my_universal_fun import checkIfNotNumeric
def selectionSort(L):
    if not(checkIfNotNumeric2(L)):
        print("Error: list does not contain numeric values")
        return
    else:        
        for i in range(len(L)):
            m,idx = findMin(L,i)
            L = swapVal(L,i,idx)
        return L
```

```python
"""Let say you are teacher and you have different student records containing 
id of a student and marks list in each subject where 
different students have taken deifferent number of subjects.
All these records are in hard copy. You want to enter all the data in the computer and
want to compute the average marks of each student and display"""

def getDataFromUser():
    DR = {}
    while True:
        key = input("Enter the id of the student :")
        if key in DR:
            print(key, "is already inserted")
        else:            
            marksList = input("Enter the marks by comma separated values: ")
            DR[key] = marksList.split(",")
        moreStudents = input("Enter yes/no for adding more students: ")
        if moreStudents.lower() == "no":
            return DR
```

```python
studentData = getDataFromUser()
```

```python
studentData
```

```python
def getAvgMarks(D):
    avgMarksDict = {}
    for x in D:
        L = D[x]
        s = 0
        for marks in L :
            s += int(marks)
        avgMarksDict[x] = s/len(L)
    return avgMarksDict
            
```

```python
avgM = getAvgMarks(studentData)
```

```python
for x in avgM:
    print("Student :",x,"got avg Marks as: ",avgM[x])
```
